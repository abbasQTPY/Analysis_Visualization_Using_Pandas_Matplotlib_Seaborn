```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

# Import seaborn and matplot libraries and load the titanic dataset

```python
titanic = sns.load_dataset('titanic')
```

# Make a barplot with sex and age 

```python
sns.barplot(x = titanic['sex'], y = titanic['age'])
```

# Find how many passengers are alive?

```python
sns.countplot(titanic["alive"])
```

# How many male and female survived w.r.t to passenger class?

```python
sns.countplot(x = titanic['pclass'],hue = titanic['sex'])
```

# Make a distribution plot of fare making the color red of the bars

```python
sns.distplot(titanic["fare"],color="red")
```

# Make a jointplot of fare and age. Write down which age group is the highest in the graph?

```python
sns.jointplot(x = titanic['age'], y = titanic['fare'])
```

# Make a boxplot of class and age?

```python
sns.boxplot(y = titanic['age'], x = titanic['class'])
```

# Show the graph of correlation between attributes of the dataset

```python
sns.heatmap(titanic.corr(), annot=True)
plt.show()
```
